# Manipulador plano de 2 ejes

```python
from IPython.display import Image
Image('2rp_new.png', width=300)
# añadir la imagen al jupyter para poder visualizarla (upload)
```

```python
# declaramos los símbolos (longitudes, variables, etc.) que se utilizarán para una formulación posterior.

from sympy.physics.mechanics import dynamicsymbols
theta1, theta2, l1, l2, theta, alpha, a, d = dynamicsymbols('theta1 theta2 l1 l2 theta alpha a d')
theta1, theta2, l1, l2, theta, alpha, a, d 
```

```python
# La matriz de transformación homogénea estándar se representa como:

rot = sp.Matrix([[sp.cos(theta), -sp.sin(theta)*sp.cos(alpha), sp.sin(theta)*sp.sin(alpha)],
                 [sp.sin(theta), sp.cos(theta)*sp.cos(alpha), -sp.cos(theta)*sp.sin(alpha)],
                 [0, sp.sin(alpha), sp.cos(alpha)]])

trans = sp.Matrix([a*sp.cos(theta),a*sp.sin(theta),d])

last_row = sp.Matrix([[0, 0, 0, 1]])

m = sp.Matrix.vstack(sp.Matrix.hstack(rot, trans), last_row)
m
```

```python
# Transformacion: 1r eje de  '0' a '1'
m01 = m.subs({alpha:0, a:l1, theta:theta1, d:0})
m01
```

```python
# Transformacion: 1r eje de '1' a '2'
m12 = m.subs({alpha:0, a:l2, theta:theta2, d:0})
m12
```

```python
# Transformación homogénea de '0' a '2'

m02 = (m01*m12)
m02
```

```python
# Utilizando los métodos de simplificación integrados de Sympy, simplificamos la matriz de transformación como

mbee= sp.Matrix([[m02[0,0].simplify(), m02[0,1].simplify(), sp.trigsimp(m02[0,3].simplify())],
                 [m02[1,0].simplify(), m02[1,1].simplify(), sp.trigsimp(m02[1,3].simplify())],
                 [m02[2,0].simplify(), m02[2,1].simplify(), m02[2,2].simplify()]])

mbee
```

Las dos primeras columnas de la matriz anterior describen la orientación de la punta del manipulador en las direcciones X, Y y Z. Considerando que, la última columna representa la posición en las direcciones respectivas. Como nuestro manipulador no tiene ningún efector final, los análisis de orientación carecen de sentido.

Ecuaciones cinemáticas donde la posición de la punta se puede expresar como:

```python
px = mbee[0,2]
print('x:')
px
```

```python
py = mbee[1,2]
print('y:')
py
```

```python
theta1, theta2, theta3, l1, l2, l3, theta, alpha, a, d = dynamicsymbols('theta1 theta2 theta3 l1 l2 l3 theta alpha a d')
theta1, theta2, theta3, l1, l2, l3, theta, alpha, a, d
```

```python
rot = sp.Matrix([[sp.cos(theta), -sp.sin(theta)*sp.cos(alpha), sp.sin(theta)*sp.sin(alpha)],
                 [sp.sin(theta), sp.cos(theta)*sp.cos(alpha), -sp.cos(theta)*sp.sin(alpha)],
                 [0, sp.sin(alpha), sp.cos(alpha)]])

trans = sp.Matrix([a*sp.cos(theta),a*sp.sin(theta),d])

last_row = sp.Matrix([[0, 0, 0, 1]])

m = sp.Matrix.vstack(sp.Matrix.hstack(rot, trans), last_row)
m
```

```python
# Transformacion: 1r eje de  '0' a '1'
m01 = m.subs({alpha:90, a:l1, theta:theta1, d:0})
m01
```

```python
m12 = m.subs({alpha:0, a:l2, theta:theta2, d:0})
m12
```

```python
m23 = m.subs({alpha:0, a:l3+theta3, theta:0, d:0})
m23
```

```python
m03 = (m01*m12*m23)
m03
```

```python
mbee= sp.Matrix([[m03[0,0].simplify(), m03[0,1].simplify(), sp.trigsimp(m03[0,3].simplify())],
                 [m03[1,0].simplify(), m03[1,1].simplify(), sp.trigsimp(m03[1,3].simplify())],
                 [m03[2,0].simplify(), m03[2,1].simplify(), m03[2,2].simplify()]])

mbee
```

```python
px = mbee[0,2]
print('x:')
px
```
